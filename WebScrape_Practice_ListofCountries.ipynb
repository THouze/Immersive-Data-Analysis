# Web Scraping Data

```python
#import the library used to query a website
import urllib.request as urllib2
```

```python
#specify the url
wiki = "https://en.wikipedia.org/wiki/List_of_counties_in_Maryland#List_of_counties"
```

```python
#Query the website and return the html to the variable 'page'
page = urllib2.urlopen(wiki)
```

```python
print(page)
```

```python
#import the Beautiful soup functions to parse the data returned from the website
from bs4 import BeautifulSoup
```

```python
#Parse the html in the 'page' variable, and store it in Beautiful Soup format
soup = BeautifulSoup(page, "lxml")
```

```python
#look at contents of page - wall of text
soup
```

```python
#format page contents to include indentation
print (soup.prettify())
```

```python
# soup.<tag>: Return content between opening and closing tag including tag.
soup.title
```

```python
# soup.<tag>.string: Return string within given tag
print(soup.title.string)
```

```python
#shows the first <a> tag on the page
soup.a
```

```python
#finds all <a> tags on the page
soup.find_all("a")
```

```python
#show hyperlink reference for all <a> tags
all_links=soup.find_all("a")
for link in all_links:
    print (link.get("href"))
```

```python
#fina all the <table> tags
all_tables=soup.find_all('table')
all_tables
```

```python
#get the <table> tag that contains the data we want to scrap
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

```python
#Generate lists
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
H=[]
I=[]
J=[]
K=[]

#

for row in right_table.findAll("tr"):
    cell = row.findAll('th')
    
    A.append(cell[0].find(text=True)) #gets info in County Name column and adds it to list A
    
    cells = row.findAll('td')
    
    if len(cells)== 10: #Only extract table body not heading
        B.append(cells[0].find(text=True)) #gets info in  FIPS column and adds it to list A
        C.append(cells[1].find(text=True)) # gets info from  County SEAT column and adds it to list B
        D.append(cells[2].find(text=True)) # gets info from ESTABLISHED column; add it to list C
        E.append(cells[3].find(text=True)) # gets info from ORIGIN column and adds it to list D
        F.append(cells[4].find(text=True)) # gets info from ENTOMOLOGY column and adds it to list E
        G.append(cells[5].find(text=True)) # gets info from FLAG column and adds it to list F
        H.append(cells[6].find(text=True)) # gets info from SEAL column and adds it to list G
        I.append(cells[7].find(text=True)) # gets info from  POPULATION column and adds it to list H
        J.append(cells[8].find(text=True)) # gets info from AREA column and adds it to list I
        K.append(cells[9].find(text=True)) # gets info from  MAP column and adds it to list J
       
```

```python
#show data in list A, just to check
A[:5]
```

```python
A.remove('County') #remove county header from list because the index was 25 and not 24 like the others which had no header
```

```python
A
```

```python
#import pandas to convert list to data frame
import pandas as pd

df=pd.DataFrame(A, columns=['County']) #turn list A into dataframe first

#can also "zip" the lists togehter and make cvs

#add other lists as new columns
df['FIPS CODE'] = B
df['County Seat'] = C
df['Established (year)'] = D
df['Origin'] = E
df['Ethymology'] = F
df['Population'] = G
df['Rank'] = H
df['Area'] = I
df['Map'] = J

df.head()
#df.print
```

```python
#export our scraped data to a file
df.to_csv("CountiesList.csv")
```
